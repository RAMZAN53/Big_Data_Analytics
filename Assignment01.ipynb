```python
import pandas as pd
import numpy as np
from google.colab import drive
```

```python
drive.mount('/content/drive')
path = "/content/drive/MyDrive/BDA/household_power_consumption.txt"
```

```python
df = pd.read_csv(path, sep='\t')
print(df)
```

```python
dataframe = df['Date;Time;Global_active_power;Global_reactive_power;Voltage;Global_intensity;Sub_metering_1;Sub_metering_2;Sub_metering_3'].str.split(';', expand=True)
dataframe.columns = ['Date', 'Time', 'Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
print(dataframe)
```

```python
rows,columns = dataframe.shape
print("Number of rows:", rows)
print("Number of columns:",columns)
```

```python
# Estimate memory usage
bytes=8
data_size = rows * columns * bytes
print(f"Estimated memory usage: {data_size / (1024**3):.2f} GB")
```

```python
# Filter the DataFrame
start_date = '2007-02-01'
end_date = '2007-02-02'
FilteredDataFrame = dataframe[(dataframe['Date'] >= start_date) & (dataframe['Date'] < end_date)]
print("Number of rows after filtering:", FilteredDataFrame.shape[0])
print("Number of columns after filtering:", FilteredDataFrame.shape[1])
```

```python
# Repplacing '?' with NaN
dataframe.replace("?", np.nan, inplace=True)
```
