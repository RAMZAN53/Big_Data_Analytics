```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive')
dataframe = pd.read_csv('/content/drive/MyDrive/BDA/dataframe.csv')
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.hist(dataframe['Global_active_power'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Global Active Power')
plt.xlabel('Global Active Power')
plt.ylabel('Frequency')
plt.grid(True)
dpi = 480 / 4.8
plt.savefig('/content/drive/MyDrive/BDA/Plot1.png', dpi=dpi)
plt.show()
```
