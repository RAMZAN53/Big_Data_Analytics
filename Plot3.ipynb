```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive')
dataframe = pd.read_csv('/content/drive/MyDrive/BDA/dataframe.csv')
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(dataframe['Date'], dataframe['Sub_metering_1'], color='skyblue', label='Sub_metering_1')
plt.plot(dataframe['Date'], dataframe['Sub_metering_2'], color='green', label='Sub_metering_2')
plt.plot(dataframe['Date'], dataframe['Sub_metering_3'], color='red', label='Sub_metering_3')
plt.title('Time Series Plot of Sub-metering Data')
plt.xlabel('Date')
plt.ylabel('Energy Sub-metering Data')
plt.legend()
plt.grid(True)
dpi = 480 / 4.8
plt.savefig('/content/drive/MyDrive/BDA/Plot3.png', dpi=dpi)
plt.show()
```
