```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drive.mount('/content/drive')
dataframe = pd.read_csv('/content/drive/MyDrive/BDA/dataframe.csv')
```

```python
plt.figure(figsize=(4.8,4.8 ))
sns.lineplot(data=dataframe['Voltage'], color='skyblue')
plt.title('Time Series Plot of Voltage')
plt.xlabel('Date')
plt.ylabel('Voltage')
plt.grid(True)
dpi = 480 / 4.8
plt.savefig('/content/drive/MyDrive/BDA/Plot4.png', dpi=dpi)
plt.show()
```
