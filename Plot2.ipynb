```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive')
dataframe = pd.read_csv('/content/drive/MyDrive/BDA/dataframe.csv')
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(dataframe['Time'], dataframe['Global_active_power'], color='black')
plt.title('Line Graph of Global Active Power')
plt.xlabel('Time')
plt.ylabel('Global Active Power')
plt.grid(True)
dpi = 480 / 4.8
plt.savefig('/content/drive/MyDrive/BDA/Plot2.png', dpi=dpi)
plt.show()
```
